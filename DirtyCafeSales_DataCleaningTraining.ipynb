```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('/content/dirty_cafe_sales.csv')
```

```python
for col in df.columns:
    print("\n")
    print("=====", col, "=====")
    print(df[col].value_counts())
```

```python
for col in df.columns:
    df = df.replace(["UNKNOWN","ERROR"], np.nan)
df.dtypes
```

```python
df[["Quantity","Price Per Unit","Total Spent"]] = df[["Quantity","Price Per Unit","Total Spent"]].astype(float)
df["Transaction Date"] = pd.to_datetime(df["Transaction Date"])
df.isnull().sum()
```

```python
quantity_filldata = (
    df["Quantity"].isnull()
    &
    df[["Price Per Unit", "Total Spent"]].notna().all(axis=1)
)

df[quantity_filldata]
```

```python
df.loc[quantity_filldata, "Quantity"] = (
    df.loc[quantity_filldata, "Total Spent"]
    / # Division
    df.loc[quantity_filldata, "Price Per Unit"]
)
```

```python
df.isnull().sum()
```

```python
price_filldata = (
    df["Price Per Unit"].isnull()
    &
    df[["Quantity","Total Spent"]].notna().all(axis=1)
)

df.loc[price_filldata, "Price Per Unit"] = (
    df.loc[price_filldata,"Total Spent"]
    /
    df.loc[price_filldata,"Quantity"]
)
```

```python
df.isnull().sum()
```

```python
spent_filldata = (
    df["Total Spent"].isnull()
    &
    df[["Quantity","Price Per Unit"]].notna().all(axis=1)
)

df.loc[spent_filldata,"Total Spent"] = (
    df.loc[spent_filldata,"Quantity"]
    *
    df.loc[spent_filldata,"Price Per Unit"]
)
```

```python
df.isnull().sum()
```

```python
quantity = df[df["Quantity"].isnull()]
quantity
```

```python
df = df.dropna(subset=["Quantity","Total Spent", "Price Per Unit"])

df.isnull().sum()
```

```python
df.shape
```

```python
df.head(10)
```

```python
temp_df = df[["Item","Price Per Unit"]].drop_duplicates(subset=["Item"])
temp_df
```

```python
items = {
    "Coffee":2,
    "Cake":3,
    'Cookie':1,
    "Salad":5,
    "Smoothie":4,
    "Sandwich":4,
    "Juice":3,
    "Tea":1.5
}

reverse_items = {
    price : item
    for item, price in items.items()
}

print(reverse_items)
```

```python
df["Item"] = df["Item"].fillna(
    df["Price Per Unit"].map(reverse_items)
)
df.isnull().sum()
```

```python
df["Payment Method"] = df["Payment Method"].fillna("Unknown")
df["Location"] = df["Location"].fillna("Unknown")
df = df.dropna(subset=["Transaction Date"])
df.isnull().sum()
```
